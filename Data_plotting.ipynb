```python
# SASA
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.rc('font',family='Times New Roman')
plt.rcParams.update({'font.size': 30})


monoli_7yim_sasa = pd.read_excel('D:/Abel/AFP_caspase/FIgure_generation/FIgure_generation/monoli_7yim_sasa.xlsx')
pinores_7yim_sasa = pd.read_excel('D:/Abel/AFP_caspase/FIgure_generation/FIgure_generation/pinores_7yim_sasa.xlsx')
syring_7yim_sasa = pd.read_excel('D:/Abel/AFP_caspase/FIgure_generation/FIgure_generation/syring_7yim_sasa.xlsx')
yim_sasa = pd.read_excel('D:/Abel/AFP_caspase/FIgure_generation/FIgure_generation/7yim_sasa.xlsx')

x1 = yim_sasa['Time (ns)']
x2 = monoli_7yim_sasa['Time (ns)']
x3 = pinores_7yim_sasa['Time (ns)']
x4 = syring_7yim_sasa['Time (ns)']
y1 = yim_sasa['Area']
y2 = monoli_7yim_sasa['Area']
y3 = pinores_7yim_sasa['Area']
y4 = syring_7yim_sasa['Area']
plt.figure(figsize=((25,15)))

plt.plot(x1, y1, label = "AFP",c='#5B2071')
plt.plot(x2, y2, label = "AFP-1-Monolinolenin complex",c='#08FF08')
plt.plot(x3, y3, label = "AFP-(-)Pinoresinol complex",c='red')
plt.plot(x4, y4, label = "AFP-(-)Syringaresinol complex",c='black')


plt.xlim(min(x3), max(x3+1))
plt.ylim(min(y2)-1, max(y2)+10)
plt.ylim(min([min(y2),min(y3),min(y4)])-100, max([max(y2),max(y3),max(y4)])+400)
plt.xlabel('Time (ns)')
plt.ylabel('SASA ()')
plt.legend(fontsize=30, loc='upper right')

plt.savefig('D:/Abel/AFP_caspase/FIgure_generation/FIgure_generation/SASA_1.png', dpi=300)
plt.show()
```
