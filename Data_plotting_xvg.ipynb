```python
##RMSD##

import matplotlib.pyplot as plt
import numpy as np
plt.rc('font',family='Times New Roman')
plt.rcParams.update({'font.size': 30})

# Loading data from the rmsd_1.xvg file, ignoring lines starting with "@" or "#"
t_control, rmsd_control = np.loadtxt("rmsd_1.xvg", comments=["@", "#"], unpack=True)

# Loading data from the rmsd_2.xvg file, ignoring lines starting with "@" or "#"
t_test, rmsd_test = np.loadtxt("rmsd_2.xvg", comments=["@", "#"], unpack=True)

# Creating a figure and axis objects for the plot
fig = plt.figure(figsize=(25,15))
ax = fig.add_subplot(111)

# Plotting the line representing the test RMSD values
ax.plot(t_test, rmsd_test * 10, color="black", linestyle="-", label="Idarubicine")  # Multiply y-coordinates by 10

# Plotting the line representing the control RMSD values
ax.plot(t_control, rmsd_control * 10, color="red", linestyle="-", label="MRTX-1133")  # Multiply y-coordinates by 10

# Setting labels for the x-axis (time in ns) and y-axis (RMSD value)
ax.set_xlabel("Time (ns)")
ax.set_ylabel(r"C$_\alpha$ RMSD (Å)")  # Using Ångstrom symbol

# Setting the x-axis and y-axis limits
ax.set_xlim(0, 200)
ax.set_ylim(0.75, 2.0)

# Adding legend
ax.legend()

# Saving the plot as a PNG image with higher resolution (300 dpi)
plt.savefig("rmsd_combined.png", format="png", dpi=300)

# Displaying the plot
plt.show()
```
